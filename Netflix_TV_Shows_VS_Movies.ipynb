#**Netflix TV shows and Movies**#

Netflix, Inc. is an American entertainment services provider in Los Gatos, California, founded in 1997 by Reed Hastings and Marc Randolph in Scotts Valley, California. As of April 2019, Netflix had over 148 million paid subscriptions worldwide, including 60 million in the United States, and over 154 million subscriptions total including free trials.

```python
# Import Libraries which are required

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
# import data directly from github

df = pd.read_csv('https://raw.githubusercontent.com/muhammadanas0716/Data-Science-Projects---EDA/main/5.%20Netflix%20TV%20Shows%20and%20Movies%20-%20Analysis/netflix_titles.csv')
```

```python
# check the data   (8807 rows × 12 columns)
df
```

```python
# check the datatype 
df.info()
```

```python
#checking number of null values in each column
df.isnull().sum()
```

```python
# replacing null value with TV ma
df['rating'].replace(np.nan,'TV-MA',inplace =True)
```

```python
# replacing null value with Unites States
df['country'].replace(np.nan,'United States',inplace =True)
```

```python
# change the column name
df.rename(columns={'listed_in':'genre'},inplace = True)
```

```python
# shape of data
df.shape
```

```python
# release date available from 1942 to 2021
df.release_year.unique()
```

```python
#value count of movies vs show
df['type'].value_counts().reset_index()
```

```python
df.head(2)
```

#The Netflix Release years, visualized via BAR GRAPH

```python
plt.rcParams['figure.figsize'] = (16,6)
plt.rcParams['figure.facecolor'] = 'white'

df['release_year'].value_counts().plot.bar(color = 'cyan',ec= 'black')
plt.xlabel('Number of releases' , fontsize = 20)
plt.ylabel('Year' , fontsize = 20)
plt.title('Number of releases in a year' , fontsize = 25)
```

#These are 2 graphs, one is a Bar Chart and the other one is a Pie Chart. This is to just visualize the amount of TV SHOWS VS MOVIES on Netflix!

```python
# TV shows and movies on netflix in Bar chart
plt.rcParams['figure.figsize'] = (10,6)
plt.rcParams['figure.facecolor'] = 'white'
plt.ylabel('Counts of releases' ,fontsize = 15)

df['type'].value_counts().plot.bar(color = ['tomato','green'],ec = 'black')   


plt.title('Number of Tv Shows and Movies' , fontsize = 20)
plt.show()
```

```python
# TV shows and movies on netflix in pie chart
plt.rcParams['figure.figsize'] = (10,6)
df['type'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

Q1: What is the proporationality between the GENRE of Movies and TV Shows?

```python
# movie genres
plt.rcParams['figure.figsize']  = (10,5)
plt.title('Top 10 Genres for Movies')
df[df['type']=='Movie']['genre'].value_counts().head(10).plot.barh(ec='black',color = sns.color_palette('pastel'))
plt.show()
```

```python
# genres of TV shows
plt.rcParams['figure.figsize']  = (10,5)
plt.title('Top 10 Genres for TV Show')
df[df['type']=='TV Show']['genre'].value_counts().head(10).plot.barh(ec='black',color = sns.color_palette('pastel'))
plt.show()
```

Q2: Who are the top 10 actors on Netflix?

```python
# remove null values from cast
no_null_cast = df[~df['cast'].isna()]
```

```python
# change str into list for using loop
type(df['cast'].str.split(',')[1])
```

```python
# create empty list
actors_list = []

# get all actors in one list
for index in no_null_cast['cast'].str.split(','):
   for i in index:
    actors_list.append(i)
```

```python
# change actors list in dataframe
actors_df = pd.DataFrame(actors_list)

# create a graph using actors df
sns.set_style('darkgrid')
plt.rcParams['figure.figsize']  = (10,5)
plt.title('Top 10 actors on netflix',fontsize = 25)

actors_df.value_counts().head(10).plot.bar(ec= 'black',color = sns.color_palette('pastel'))
plt.xlabel('Actors',fontsize = 15)
plt.show()
```

Q3: Does Netflix upload more Movies or TV shows?

```python
plt.figure(figsize=(15,8))
plt.plot(df[df['type']=='Movie'].groupby('release_year')['type'].count(),color='salmon',linewidth = 4)
plt.plot(df[df['type']=='TV Show'].groupby('release_year')['type'].count(), color='lightblue',linewidth = 4)
plt.xlabel("Year",fontsize = 15)
plt.ylabel("Titles Added",fontsize = 15)
plt.title("Movies vs. TV Shows on Netflix\nCount of Titles Released per Year",fontsize = 20)
plt.legend(['TV Shows','Movies'])
plt.show()
```

Q4: When is the Best to time to release a Movie/TV shows?

This heatmap shows the best time of the year (WHEN IT WAS) to release to Netflix, I.E Month and Year. Moreover you can see that the best time to release looks like March, July, Auguest and September. You can actually use this to predict future results. In my opinion I see a very nice trend in July as the heatmap is getting darker as we move rightwards!

```python
# drop all null values from date added column
no_null_date =df[['date_added']].dropna()

# change datatypeof date added column from object to date time
no_null_date['date_added'] = no_null_date['date_added'].astype('datetime64') 
```

```python
# extract year and months 
no_null_date['year'] = no_null_date['date_added'].dt.year
no_null_date['month'] = no_null_date['date_added'].dt.month
```

```python
no_null_date
```

```python
# grouped with respect to year 
df_new = no_null_date[['year','month']].groupby('year')['month'].value_counts().unstack().fillna(0)

# create a heatmap
plt.figure(figsize=(15,8))
colormap = sns.color_palette("Reds")
sns.heatmap(df_new,cmap = colormap)
plt.title('Best time to release to netflix',fontsize = 20)
plt.show()
```

Q5: What are the TV shows with largest number of seasons?

```python
# create a df which have only Tv show only
TV_show_df = df[df['type']=='TV Show']

sns.set_style('darkgrid')
plt.rcParams['figure.figsize']  = (15,7)
plt.title('TV shows with largest number of seasons',fontsize = 25)

# counts all the values of duration(seasons)
TV_show_df['duration'].value_counts().plot.bar(ec='black',color = sns.color_palette('RdYlBu'))
plt.show()
```

Q6:TV shows with maximum number of seasons

```python
# convert duration column into int
TV_show_df['duration'] = TV_show_df['duration'].astype('int64')
```

```python
sns.set_style('darkgrid')
plt.rcParams['figure.figsize']  = (15,7)
plt.title('TV shows with max number of seasons',fontsize = 25)

# sort values with duration for getting max no of seasons 
my_df = TV_show_df[['title','duration']].sort_values('duration',ascending =False).reset_index(drop = 'index').head(20)

# create barplot
sns.barplot(y=my_df['title'] ,x= my_df['duration'])
plt.show()
```

Q6: Who are the most famous directors?

```python
# crate a copy of datframe for doing any changes
df_dir = df.copy()
```

```python
# remove null values from director column 
df_dir = df_dir.dropna(subset=['director'])
```

```python
# create a horizontal bar chart
sns.set_style('darkgrid')
plt.rcParams['figure.figsize']  = (15,7)
plt.title('Most Famous Directors in movies',fontsize = 25)
df_dir[df_dir['type'] == 'Movie']['director'].value_counts().head(15).plot.barh(ec ='black',color = sns.color_palette('RdYlBu'))
plt.show()
```

```python
plt.title('Most Famous Directors in TV shows',fontsize = 25)
df_dir[df_dir['type'] == 'TV Show']['director'].value_counts().head(10).plot.barh(ec ='black',color = sns.color_palette('RdYlBu'))
plt.show()
```
